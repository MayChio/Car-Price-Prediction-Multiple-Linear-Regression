```python
#Importar librerÃ­as necesarias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Cargar el dataset
df = pd.read_csv("CarPrice_Assignment.csv")

# ðŸ‘€ Ver las primeras filas
df.head()
```

```python
# GrÃ¡fico de dispersiÃ³n
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='horsepower', y='price')
plt.title('RelaciÃ³n entre Horsepower y Price')
plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
# Variables para regresiÃ³n
X = df[['horsepower']]  # Variable independiente
y = df['price']         # Variable dependiente

# Crear y entrenar el modelo
modelo = LinearRegression()
modelo.fit(X, y)

# Predicciones
y_pred = modelo.predict(X)

# Resultados del modelo
print("Coeficiente (pendiente):", modelo.coef_[0])
print("IntersecciÃ³n:", modelo.intercept_)
print("RÂ² (score):", modelo.score(X, y))
```

```python
#GrÃ¡fica con lÃ­nea de regresiÃ³n
plt.figure(figsize=(8,5))
sns.scatterplot(x=X['horsepower'], y=y, label='Datos reales')
plt.plot(X, y_pred, color='red', label='LÃ­nea de regresiÃ³n')
plt.title('RegresiÃ³n Lineal: Horsepower vs Price')
plt.xlabel('Horsepower')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# GrÃ¡fico de dispersiÃ³n
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='enginesize', y='price')
plt.title('RelaciÃ³n entre Engine Size y Price')
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

```python
# GrÃ¡fico con lÃ­nea de regresiÃ³n
plt.figure(figsize=(8,5))
sns.scatterplot(x=X2['enginesize'], y=y2, label='Datos reales')
plt.plot(X2, y2_pred, color='red', label='LÃ­nea de regresiÃ³n')
plt.title('RegresiÃ³n Lineal: Engine Size vs Price')
plt.xlabel('Engine Size')
plt.ylabel('Price')
plt.legend()
plt.grid(True)
plt.show()
```
